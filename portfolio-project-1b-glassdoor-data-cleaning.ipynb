# Removing Number in Company Name columns

# Salary Estimate Column

* Check Values in ('Salary Estimate').column
* Create New columns (min salary,max salary,avg salary)
* Remove Extra symbols And character in Values

```python
uncleaned_data['Salary Estimate'].value_counts()
```

```python
uncleaned_data['min_salary']=0
uncleaned_data['max_salary']=0
uncleaned_data['avg_salary']=0
for i in range(len(uncleaned_data)):
    try:
        uncleaned_data.loc[i,"min_salary"]=int(uncleaned_data['Salary Estimate'][i].split(" ")[0].split("-")[0].replace("$","").replace("K",""))
        uncleaned_data.loc[i,"max_salary"]=int(uncleaned_data['Salary Estimate'][i].split(" ")[0].split("-")[1].replace("$","").replace("K",""))
    except:
        uncleaned_data.loc[i,"min_salary"]=int(uncleaned_data['Salary Estimate'][i].split("(E")[0].split("-")[0].replace("$","").replace("K",""))
        uncleaned_data.loc[i,"max_salary"]=int(uncleaned_data['Salary Estimate'][i].split("(E")[0].split("-")[1].replace("$","").replace("K",""))
    finally:
        uncleaned_data.loc[i,"Salary Estimate"]=str(uncleaned_data.loc[i,"min_salary"])+"-"+str(uncleaned_data.loc[i,"max_salary"])
        uncleaned_data.loc[i,"avg_salary"]=np.mean([uncleaned_data.loc[i,"min_salary"],uncleaned_data.loc[i,"max_salary"]])
```

# Exctract Data Job Description
* Create New Columns Based on Skill required for Job (Python,Excel,Sql,Tableau,Sparx,mechine learing,Aws) If Job description Contain Those skill in column print '1' otherwise '0' will print

```python
uncleaned_data['Job Description'][0].split("\n\n")
```

```python
uncleaned_data['python']=uncleaned_data['Job Description'].apply(lambda x: 1 if "python" in x.lower() else 0)
uncleaned_data['excel']=uncleaned_data['Job Description'].apply(lambda x: 1 if "excel" in x.lower() else 0)
uncleaned_data['Sql']=uncleaned_data['Job Description'].apply(lambda x: 1 if "sql" in x.lower() else 0)
uncleaned_data['tableau']=uncleaned_data['Job Description'].apply(lambda x: 1 if "tableau" in x.lower() else 0)
uncleaned_data['Sparx']=uncleaned_data['Job Description'].apply(lambda x: 1 if "Sparx" in x.lower() else 0)
uncleaned_data['machine learing']=uncleaned_data['Job Description'].apply(lambda x: 1 if "mechine learing" in x.lower() else 0)
uncleaned_data['aws']=uncleaned_data['Job Description'].apply(lambda x: 1 if "aws" in x.lower() else 0)
```

# Create New Column State Useing ['Location'] column

```python
uncleaned_data['Location'].apply(lambda x:x.split(",")[-1]).value_counts()
```

```python
uncleaned_data['State']= uncleaned_data['Location'].apply(lambda x:x.split(",")[-1].strip())
uncleaned_data['State'].replace(['United States','New Jersey','California','Texas ','Remote','Utah'],["US","NJ","CA","Tx",np.nan,np.nan],inplace=True)
uncleaned_data.dropna(axis=0,inplace= True)
uncleaned_data.reset_index(inplace= True)
```

# Create New Colum Employeetype Based on ['Job Titile']
* Based on the job title tag like sr, senior,lead,vp...etc find the employee is Senior or junior

```python
uncleaned_data['Job Title'].value_counts()
```

```python
def employeetype(job_title):
    job_title=job_title.lower()
    emp=['sr','senior','lead','principal','vp','vice president','director']
    for i in emp:
        if i in job_title:
            return "Senior_emp"
        else:
            return "junior_emp"
    return NaN
```

```python
uncleaned_data['Employee expriece ']= uncleaned_data['Job Title'].apply(employeetype)
uncleaned_data
```
